```python
# Load data
import pandas as pd

df = pd.read_excel("marked_headlines.xlsx")
len(df)
```

```python
# delete duplicated titles
df = df[~df['Title'].duplicated()]
print(len(df))
```

```python
import string

# QA: Should I remove '-'?
# QA: Should I separate '2022.gada' or leave it as '2022gada' ?
# TODO: check result more carefully
def remove_punctuation(title):
    text = "".join([char for char in title.lower() if char not in string.punctuation])
    return text


df['no_punct_title'] = df['Title'].apply(lambda x: remove_punctuation(x))
df
```

```python
import re

def tokenise(title):
    # split by 1 or more spaces
    return re.split('\s+', title)

df['title_tokens'] = df['no_punct_title'].apply(lambda x: tokenise(x))
df
```

```python
def remove_stopwords(title):
    return [token for token in title if token not in stopwords.stopwords("lv")]


df['no_stopw_title'] = df['title_tokens'].apply(lambda x: remove_stopwords(x))
df
```

```python
df.to_excel('stopwords-revomed.xlsx', index=False)
```

### Lemmatizēšana lietojot Stanza

```python
# Load data
import pandas as pd

df = pd.read_excel("out.xlsx")  # stopwords-revomed
len(df)
```

```python
df.columns
```

```python
import stanza

# 1. download lv model
stanza.download("lv")
nlp = stanza.Pipeline("lv", processors="tokenize,lemma", tokenize_no_ssplit=True)

# 2.  lemmatization
def lemmatize(sentence: str):
    doc = nlp(sentence)
    return [w.lemma for w in doc.iter_words()]
```

```python
df["lemma"] = df["no_punct_title"].apply(lambda x: lemmatize(x))
df.head(5)
```

```python
df.to_excel('out.xlsx', index=False)
```

```python
# Check how much words stay unchangable
df_posible_errors = pd.DataFrame()

def suspicious_words(sentence: str):
    doc = nlp(sentence)
    return [w.text for w in doc.iter_words() if w.text == w.lemma]

df_posible_errors["suspicious"] = df["no_punct_title"].apply(suspicious_words)
df_posible_errors
```

```python
df_posible_errors.to_excel('possible_errors.xlsx', index=False)
```

```python
# Load data
import pandas as pd

df = pd.read_excel('marked_headlines.xlsx')
len(df)
```

```python
# delete duplicated titles
df = df[~df['Title'].duplicated()]
print(len(df))
```

```python
df.columns
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm
import numpy as np

# load model
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")
model = AutoModel.from_pretrained("xlm-roberta-base")
model.eval()  # dropout off

def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1)
    return embedding.squeeze().numpy()


embeddings = []
for sentence in tqdm(df["Title"], desc="Encoding"):
    sent_vec = get_embedding(sentence)
    embeddings.append(sent_vec)


# Create an dataFrame (size: 768)
embeddings_df = pd.DataFrame(embeddings)
embeddings_df.columns = [f"xlmr_{i}" for i in range(embeddings_df.shape[1])]

# Concat to one dataframe
df_with_embeddings = pd.concat([df, embeddings_df], axis=1)

print(df_with_embeddings.head())
print(df_with_embeddings.columns)
```

```python
# Concat to one dataframe
df_with_embeddings = pd.concat([df, embeddings_df], axis=1)
print(df_with_embeddings.head())
print(df_with_embeddings.columns)
```
