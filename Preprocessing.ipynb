```python
# Load data
import pandas as pd

df = pd.read_excel("marked_headlines.xlsx")
len(df)
```

```python
# delete duplicated titles
df = df[~df['Title'].duplicated()]
print(len(df))
```

```python
import string

# QA: Should I remove '-'?
# QA: Should I separate '2022.gada' or leave it as '2022gada' ?
# TODO: check result more carefully
def remove_punctuation(title):
    text = "".join([char for char in title.lower() if char not in string.punctuation])
    return text


df['no_punct_title'] = df['Title'].apply(lambda x: remove_punctuation(x))
df
```

```python
import re

def tokenise(title):
    # split by 1 or more spaces
    return re.split('\s+', title)

df['title_tokens'] = df['no_punct_title'].apply(lambda x: tokenise(x))
df
```

```python
def remove_stopwords(title):
    return [token for token in title if token not in stopwords.stopwords("lv")]


df['no_stopw_title'] = df['title_tokens'].apply(lambda x: remove_stopwords(x))
df
```

```python
df.to_excel('stopwords-revomed.xlsx', index=False)
```

### Lemmatizēšana lietojot Stanza

```python
# Load data
import pandas as pd

df = pd.read_excel("out.xlsx")  # stopwords-revomed
len(df)
```

```python
df.columns
```

```python
import stanza

# 1. download lv model
stanza.download("lv")
nlp = stanza.Pipeline("lv", processors="tokenize,lemma", tokenize_no_ssplit=True)

# 2.  lemmatization
def lemmatize(sentence: str):
    doc = nlp(sentence)
    return [w.lemma for w in doc.iter_words()]
```

```python
df["lemma"] = df["no_punct_title"].apply(lambda x: lemmatize(x))
df.head(5)
```

```python
df.to_excel('out.xlsx', index=False)
```

```python
# Check how much words stay unchangable
df_posible_errors = pd.DataFrame()

def suspicious_words(sentence: str):
    doc = nlp(sentence)
    return [w.text for w in doc.iter_words() if w.text == w.lemma]

df_posible_errors["suspicious"] = df["no_punct_title"].apply(suspicious_words)
df_posible_errors
```

```python
df_posible_errors.to_excel('possible_errors.xlsx', index=False)
```

```python
# Load data
import pandas as pd

df = pd.read_excel('out.xlsx')
len(df)
```

```python
df.columns
```

```python
# concat lemmas array to the sentence of lemmas
import ast

# Important fix that converts <list representation in string type> to actual list type
# Lemmatizator result somehow return list representation in string.
df["lemma"] = df["lemma"].apply(ast.literal_eval)
# Convert list of words into string(sentence)
df['lemma_sentence'] = df['lemma'].apply(" ".join)
df.head(5)                                
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm
import numpy as np

# load model
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")
model = AutoModel.from_pretrained("xlm-roberta-base")
model.eval()  # dropout off

def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)

        # [batch_size, seq_len, hidden_size]
        token_embeddings = outputs.last_hidden_state.squeeze(0)  # remove batch dim → shape: [seq_len, hidden_size]

        # create resulting meaned vector of the each token embedding
        sentence_embedding = token_embeddings.mean(dim=0)  # shape: [hidden_size]

    return sentence_embedding.numpy(), token_embeddings.numpy()


sentence_embeddings = []
token_embeddings = []
for lemma_sentence in tqdm(df["lemma_sentence"], desc="Encoding"):
    sent_vec, token_vec = get_embedding(lemma_sentence)
    sentence_embeddings.append(sent_vec)
    token_embeddings.append(token_vec)


# Create an dataFrame (size: 768) and save np.array
sentence_embeddings_df = pd.DataFrame(sentence_embeddings)
sentence_embeddings_df.columns = [f"xlmr_{i}" for i in range(sentence_embeddings_df.shape[1])]
token_embeddings_np = np.array(token_embeddings, dtype=object)

# Concat to one dataframe
df_with_embeddings = pd.concat([df, sentence_embeddings_df], axis=1)

print(df_with_embeddings.head())
print(token_embeddings_np[:2])
```

```python
sentence_embeddings_df.to_parquet("sentence_embeddings_only.parquet")
df_with_embeddings.to_parquet("sentence_embeddings_with_headlines_marks_etc.parquet")
# To retrieve:
# df = pd.read_parquet("name.parquet")
np.save("token_embeddings.npy", token_embeddings_np)
print('Saved!')
```

```python
# Start session

import sparknlp

from sparknlp.base import *
from sparknlp.annotator import *

from pyspark.ml import Pipeline

spark = sparknlp.start()
```

```python
# pipeline for getting Lemma

document = DocumentAssembler().setInputCol("text").setOutputCol("document")

sentence = SentenceDetectorDLModel.pretrained("sentence_detector_dl", "xx").setInputCols(["document"]).setOutputCol("sentence")

token = Tokenizer().setInputCols(["sentence"]).setOutputCol("token") 

lemma = LemmatizerModel.pretrained("lemma_lvtb", "lv").setInputCols(["token"]).setOutputCol("lemma")

embeddings = RoBertaEmbeddings.pretrained("roberta_base_latvian_legal", "lv")\
  .setInputCols(["lemma"])\
  .setOutputCol("embeddings")

pipeline = Pipeline(stages=[document, sentence, token, lemma, embeddings])



data = spark.createDataFrame([['trūkstošu'],['izziņu'],[ 'negatīviem'],
                              ['testiem'], ['pagājušajā'], ['nedēļā'], ['lidojumi'],
                              ['rīgu'], ['atteikti'], ['personai']]).toDF("text")

result = pipeline.fit(data).transform(data)
result.show(5)
```
